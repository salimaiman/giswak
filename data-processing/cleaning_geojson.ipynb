# CLEANING GEOJSON FILE

```python
import pandas as pd
import geopandas as gpd
```

```python
gdf = gpd.read_file('ntt.geojson')
gdf.head()
```

```python
gdf['NAME_2'] = gdf['NAME_2'].replace({
    'FloresTimur': 'Flores Timur',
    'ManggaraiBarat': 'Manggarai Barat',
    'ManggaraiTimur': 'Manggarai Timur',
    'KotaKupang': 'Kota Kupang',
    'RoteNdao': 'Rote Ndao',
    'SabuRaijua': 'Sabu Raijua',
    'SumbaBarat': 'Sumba Barat',
    'SumbaBaratDaya': 'Sumba Barat Daya',
    'SumbaTengah': 'Sumba Tengah',
    'SumbaTimur': 'Sumba Timur',
    'TimorTengahSelatan': 'Timor Tengah Selatan',
    'TimorTengahUtara': 'Timor Tengah Utara'
})
```

```python
gdf.head()
```

```python
gdf.to_file('ntt.geojson', driver='GeoJSON')
```
