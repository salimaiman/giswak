### REKAP SEDERHANA

```python
kabupaten = df['KABUPATEN'].unique()
rekap_df = pd.DataFrame(columns=['KABUPATEN', 'JUMLAH BIDANG', 'TOTAL LUAS (M2)', 'SUDAH BERSERTIFIKAT', 'BELUM BERSERTIFIKAT', 'PERSENTASE (%)'])
rekap_df['KABUPATEN'] = kabupaten
rekap_df['KABUPATEN'] = rekap_df['KABUPATEN'].replace('KAB.KUPANG', 'Kupang')
rekap_df['KABUPATEN'] = rekap_df['KABUPATEN'].replace('KAB. SIKKA', 'Sikka')
```

```python
counts = []
for kab in kabupaten:
    count = df[df['KABUPATEN'] == kab].shape[0]
    counts.append(count)

rekap_df['JUMLAH BIDANG'] = counts
```

```python
luas = []
for kab in kabupaten:
    total_luas = df[df['KABUPATEN'] == kab]['LUAS'].sum()
    luas.append(f"{total_luas}")
rekap_df['TOTAL LUAS (M2)'] = luas
```

```python
sudah_sertifikat = []
for kab in kabupaten:
    count = df[(df['KABUPATEN'] == kab) & (df['STATUS SERTIFIKAT'] == 'SUDAH')].shape[0]
    sudah_sertifikat.append(count)
rekap_df['SUDAH BERSERTIFIKAT'] = sudah_sertifikat
```

```python
belum_sertifikat = []
for kab in kabupaten:
    count = df[(df['KABUPATEN'] == kab) & (df['STATUS SERTIFIKAT'] == 'BELUM')].shape[0]
    belum_sertifikat.append(count)
rekap_df['BELUM BERSERTIFIKAT'] = belum_sertifikat
```

```python
persentase = []
for i in range(len(rekap_df)):
    if rekap_df.loc[i, 'JUMLAH BIDANG'] > 0:
        persentase_value = (rekap_df.loc[i, 'SUDAH BERSERTIFIKAT'] / rekap_df.loc[i, 'JUMLAH BIDANG'])
    else:
        persentase_value = 0
    persentase.append(f"{persentase_value:.2f}")
rekap_df['PERSENTASE (%)'] = persentase
```

```python
rekap_df['KABUPATEN'] = rekap_df['KABUPATEN'].str.title()
df['KABUPATEN'] = df['KABUPATEN'].str.title()
```

```python
kat_mushalla = ['Langgar', 'Mushalla', 'Musala', 'Musollah', 'Mushola', 'Musholla', 'Musholah', 'Langgar/ Mushalla', 'Langgar/Mushalla', 'Langgar/Mushola', 'Langgar/Musholla', 'Langgar/Musala', 'Langgar/Musollah', 'Langgar/Musollah', 'Langgar/Musolah']
kat_masjid = ['Masjid', 'Masjid Jami', 'Masjid Jami\'', 'Masjid Raya', 'Masjid Agung']
kat_sekolah = ['Sekolah', 'Madrasah', 'Pesantren', 'Pondok Pesantren', 'Pondok Pesantren/Madrasah', 'Pesantren/Madrasah']
kat_makam = ['Perkuburan', 'Perkuburan Islam' , 'Kuburan', 'Makam', 'Makam Islam', 'Makam/Kuburan', 'Makam/Perkuburan', 'Kuburan/Makam', 'Kuburan/Perkuburan', 'Perkuburan/Makam', 'Perkuburan/Kuburan']
kat_hijau = ['Kebun', 'Taman', 'RTH', 'Ruang Terbuka Hijau', 'RTH/Taman', 'Taman/RTH']
def categorize_peruntukkan(value):
    if pd.isna(value):
        return 'Lainnya'
    value = value.strip().lower()
    if any(kat.lower() in value for kat in kat_mushalla):
        return 'Mushalla'
    elif any(kat.lower() in value for kat in kat_masjid):
        return 'Masjid'
    elif any(kat.lower() in value for kat in kat_sekolah):
        return 'Sekolah'
    elif any(kat.lower() in value for kat in kat_makam):
        return 'Makam'
    elif any(kat.lower() in value for kat in kat_hijau):
        return 'Ruang Terbuka Hijau'
    else:
        return 'Kemaslahatan Umat'
    
df['KATEGORI'] = df['PERUNTUKKAN PADA AIW'].apply(categorize_peruntukkan)
```

```python
df['PEMANFAATAN SAAT INI'] = df['PEMANFAATAN SAAT INI'].fillna(df['PERUNTUKKAN PADA AIW'])
```

```python
df['PERUNTUKKAN PADA AIW'] = df['PERUNTUKKAN PADA AIW'].fillna('Masjid')
```

```python
df = df.fillna('-')
```

```python
df['KABUPATEN'] = [kab.upper() for kab in df['KABUPATEN']]
```

```python
df['KABUPATEN'] = df['KABUPATEN'].str.title()
rekap_df['KABUPATEN'] = rekap_df['KABUPATEN'].str.title()
```

```python
df.columns = [col.lower().replace(' ', '_') for col in df.columns]
df = df.rename(columns={'koordinat': 'coordinate'})
rekap_df.columns = [col.lower().replace(' ', '_') for col in rekap_df.columns]
```

### EXPORT TO GPX

```python
df
```

```python
df_gpx = df[['kabupaten','pemanfaatan_saat_ini', 'coordinate']]
df_gpx = df_gpx[df_gpx['kabupaten'] ==  'Kota Kupang']
df_gpx = df_gpx.rename(columns={'pemanfaatan_saat_ini': 'name'})
df_gpx['latitude'] = df_gpx['coordinate'].apply(lambda x: float(x.split(',')[0]) if pd.notna(x) and ',' in x else None)
df_gpx['longitude'] = df_gpx['coordinate'].apply(lambda x: float(x.split(',')[1]) if pd.notna(x) and ',' in x else None)
df_gpx = df_gpx.drop(columns=['coordinate'])
df_gpx = df_gpx.dropna(subset=['latitude', 'longitude'])
```

```python
df_gpx.head()
```

```python
import gpxpy
import gpxpy.gpx

# Create GPX object
gpx = gpxpy.gpx.GPX()

# Create a GPX track
track = gpxpy.gpx.GPXTrack()
gpx.tracks.append(track)

# Create a segment in the track
segment = gpxpy.gpx.GPXTrackSegment()
track.segments.append(segment)

# Loop through DataFrame rows and add points
for _, row in df_gpx.iterrows():
    point = gpxpy.gpx.GPXTrackPoint(
        latitude=row['latitude'],
        longitude=row['longitude'],
        name=row.get('name', None)
    )
    segment.points.append(point)

# Save to file
with open('kota_kupang.gpx', 'w') as f:
    f.write(gpx.to_xml())

print("✅ GPX file created: output.gpx")
```
